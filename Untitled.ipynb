```python
def load_graph(state: str):
    gdf = gpd.read_file(f"/home/max/git/vtd-migration/products/{state}_vtd20.shp")
    graph = Graph.from_geodataframe(gdf)
    graph.to_json(f"graphs/{state}_vtd20.json")
    
    counties, nodes_by_county = get_divisions(graph, 'COUNTYFP20')
    return graph, counties, node_by_county
```

```python
def get_district_information(partition, group):
    group_percents = {d:partition[group][d]/partition["VAP20"][d] for d in partition[group].keys()}
    majorities = len([True for item in group_percents.values() if item >= 0.5])

    # assumes that there must exist a district below 0.5
    next_highest = max([item for item in group_percents.values() if item < 0.5])

    return majorities + next_highest

def majority_acceptance_factory(group, exp=1):
    def search_for_multiple_majority_districts(partition):
        score = get_district_information(partition, group)
        parent_score = get_district_information(partition.parent, group)

        if score > parent_score:
            return True
        else:
            probability = (score/parent_score)**exp

            if random.random() < probability:
                return True
            else:
                return False
    return search_for_multiple_majority_districts
```

```python
def run_short_burst(state, bursts, burst_length):
    global best_score
    global splits
    global number_of_county_splits
    global best_partition
    
    graph, counties, node_by_county = create_graph(state)
    TOTPOP = 'TOTPOP20'
    num_dists = 7
    epsilon = 0.02
    ideal_population = sum([graph.nodes[n]['TOTPOP20'] for n in graph.nodes]) / num_dists

    result = recursive_seed_part(
        graph,
        range(num_dists),
        ideal_population,
        TOTPOP,
        epsilon,
        method=bipartition_tree,
        n = None,
        ceil = None
    )
    
    updaters = {}
    updaters["population"] = Tally(TOTPOP, alias="population")
    updaters['VAP20'] = Tally('VAP20')
    updaters['BVAP20'] = Tally('BVAP20')
    
    seed_partition = GeographicPartition(
        graph=graph,
        assignment=result,
        updaters=updaters
    )

    
    proposal = partial(
        recom,
        method = partial(division_bipartition_tree,
                division_tuples=[('COUNTYFP20', 1)],
                first_check_division=True),
        pop_col=TOTPOP,
        pop_target=ideal_population,
        epsilon=epsilon,
        node_repeats=2
    )
    graph, counties, node_by_county = create_graph("VA")
    constraint = constraints.within_percent_of_ideal_population(best_partition, epsilon)
    for i in range(bursts):
        chain = MarkovChain(
            proposal = proposal,
            constraints = [constraint],
            accept= majority_acceptance_factory('BVAP20',3),
            initial_state = best_partition,
            total_steps = burst_length
        )
        for part in chain:
            score = get_district_information(part, 'BVAP20')
            scores.append(score)
            print(f"score: {score}, best score: {best_score}")
            if score > best_score:
                best_score = score
                best_partition = part

            splits = num_division_splits(graph,part,counties,nodes_by_county)
            number_of_county_splits.append(splits)

            yield part
```
